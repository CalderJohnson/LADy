# **Quickstart**


For quickstart purposes, a toy sample of reviews has been provided at ./data/raw/semeval/toy.2016SB5/ABSA16_Restaurants_Train_SB1_v2.xml. You can run LADy by:

```python
%cd ..
# %cd src
```

```python
!python main.py -naspects 5 -am rnd -data ../data/raw/semeval/toy.2016SB5/ABSA16_Restaurants_Train_SB1_v2.xml -output ../output/toy.2016SB5/ABSA16_Restaurants_Train_SB1_v2.xml/
```



---



where the input arguements are:

`aml`: A list of aspect modeling methods among {`lda`, `rnd`, `btm`}, required, case-insensitive.
  
`data`: Raw dataset file path, required.
  
`output`: Output path, required.
  
`naspects`: Number of aspects, required.
